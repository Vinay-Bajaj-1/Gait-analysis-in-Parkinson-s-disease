```python
#combine 4 text files into 1 
for j in range(1, 13):
    if j == 8:
        print(j, end = " ")
        main_df = pd.DataFrame(columns = columns)
        for k in range(1,3):
            directory = 'C:\\Users\\Vinay\Gait Analysis\\Filtered Data\\008\\OFF_' + str(k)
            num_files = len(os.listdir(directory))

            for i in range(1,num_files + 1):
                df = pd.read_csv(directory+ '\\task_' + str(i) +'.txt', header = None).iloc[:, 2:]
                df.insert(0, 'person', j)
                df.columns = columns
                main_df = pd.concat([main_df, df], ignore_index = True)
                
        print(main_df.shape)
        main_df.to_csv('C:\\Users\\Vinay\\Gait Analysis\\merged files main\\' + str("{:03d}".format(j) + '.csv'), index = False)
        continue

    else:
        main_df = pd.DataFrame(columns = columns)
        directory = "C:\\Users\\Vinay\Gait Analysis\\Filtered Data\\" + str("{:03d}".format(j))
        num_files = len(os.listdir(directory))
        print(j, end = " ")

        for i in range(1,num_files+1):
            df = pd.read_csv(directory+ '\\task_' + str(i) +'.txt', header = None).iloc[:, 2:]
            df.insert(0, 'person', j)
            df.columns = columns

            main_df = pd.concat([main_df, df], ignore_index = True)
        main_df.to_csv('C:\\Users\\Vinay\\Gait Analysis\\merged files main\\' + str("{:03d}".format(j) + '.csv'), index = False)
        print(main_df.shape)
        
        
```
